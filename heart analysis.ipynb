```python
heart_df['num'].value_counts()
```

```python
# how many people have heart disease and how many dont in this dataset
sns.countplot(x='num',data=heart_df)
plt.title('count of eart disease cases')
plt.xticks([0,1],['absence','presence'])
plt.show()
```

```python
heart_df['sex'].value_counts()
```

```python
# count of male and female in this dataset
sns.countplot(x='sex',data=heart_df)
plt.xticks([0,1],['female','male'])
plt.show()
```

```python
# gender distribution according to the num variables
sns.countplot(x='sex',hue='num',data=heart_df)
plt.xticks([0,1],['female','male'])
plt.legend(labels=['no disease','disease'])
plt.show
```

```python
#to check the age distribution in the dataset
sns.histplot(x='age',data=heart_df,bins=20,kde=True)
plt.show()
```

```python
# to check chest pain type
sns.countplot(x='cp',data=heart_df)
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal pain','asymptomatic'])
plt.show()
```

```python
# chest pain distribution as per num variable
sns.countplot(x='cp',hue='num',data=heart_df)
plt.legend(labels=['no disease','disease'])
plt.show()
```

```python
# fast blood sugar distribution according to num variable
sns.countplot(x='fbs',hue='num',data=heart_df)
plt.legend(labels=['no disease','disease'])
plt.show()
```

```python
#resting blood pressure distribution
heart_df['trestbps'].hist()
```

```python
# comparing resting blood pressure as per sex column
g = sns.FacetGrid(heart_df,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend(labels=['male','female'])
plt.show()
```
